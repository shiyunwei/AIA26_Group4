## Variables 變數
> **Definition of Variables: used to represent a certain value.**


```python
>>> a = 1   # Declare a variable named 'a' and assign it the value 1 — this action is called "assignment"
>>> b = 2   # Similarly, create a variable named 'b'
>>> a       # Then, print the value that the variable 'a' refers to
1
>>> b
2
>>>
```

We can use this diagram to understand the relationship between variables and values:

![Variable diagram](https://i.imgur.com/wyocCHg.png)

We can also change the value that a variable points to:

```python
>>> a = 2    # Let 'a' point to 2 instead
>>> a
2
>>>
```

Now, our diagram looks like this…

![Variable diagram](https://i.imgur.com/zVUrVmE.png)


We can make one variable (b) point to another variable (a),
which means b will also point to the value that a currently points to:

```python
>>> a = 1
>>> b = a  # This makes 'b' take on the value that 'a' currently has, which is 1
>>> a = 5
>>> b      # The value of 'b' does not change; it remains 1
1
>>>
```

If we use a variable that hasn’t been declared, an error will occur:

```python
>>> thingy
Traceback (most recent call last):
  File "<stdin>", line 1, in <module>
NameError: name 'thingy' is not defined
>>>
```


- A variable always points to a value — it never points to another variable.
- Multiple different variables can point to the same value, but one variable cannot point to multiple values.


In Python, variables are case-sensitive, which means that uppercase and lowercase letters are treated as different characters:

```python
>>> thing = 1
>>> THING = 2
>>> thIng = 3
>>> thing
1
>>> THING
2
>>> thIng
3
>>>
```

These keywords have special meanings and will be explained in more detail later.
> You can type `help('keywords')` in Python to see a full list of reserved words.

When we use `=` for assignment, note that the expression on the right side of `=` is evaluated first.
This means we can perform calculations on the right-hand side before assigning the result to the variable.

```python
>>> a = 1
>>> a = a + 1
```

```python

>>> a *= 2          # a = a * 2
```

## Other Comparison Operators

So far, we’ve only learned about `==`, but in Python there are many other ways to compare values.


|   usage   | Description                       | True examples         |
|-----------|-----------------------------------|-----------------------|
| `a == b`  | a is equal to b                   | `1 == 1`              |
| `a != b`  | a is not equal to b               | `1 != 2`              |
| `a > b`   | a is greater than b               | `2 > 1`               |
| `a >= b`  | a is greater than or equal to b   | `2 >= 1`, `1 >= 1`    |
| `a < b`   | a is less than b                  | `1 < 2`               |
| `a <= b`  | a is less than or equal to b      | `1 <= 2`, `1 <= 1`    |


We can use `and` and `or` to extend a longer caculation.

| Usage     | Description                               | True example                      |
|-----------|-------------------------------------------|-----------------------------------|
| `a and b` | a is True and b is True                   | `1 == 1 and 2 == 2`               |
| `a or b`  | a is True, b is True or they're both True | `False or 1 == 1`, `True or True` |




`not` means the opposite (logical negation).

```python
>>> not False
True
>>> not True
False
```

## indention 縮排
Python uses indentation (spaces) to indicate different code blocks.
```python
if something:
    print('123123')
    print('testing')
else:
    print('else something')
    print('same block in testing')
```
In the example above, if the if statement evaluates to True, the program executes lines 3 and 4, because they share the same indentation level.
Similarly, if the else statement is true, the program executes lines 5 and 6.



## Using if statements

The purpose of using if statements is to perform conditional logic,
which follows the pattern: `if...then...`.
In Python, it allows your program to make decisions and execute specific code only when certain conditions are met.


```python
>>> its_raining = True
>>> if its_raining:
...     print("It's raining!")
...
It's raining!
>>> its_raining = False
>>> if its_raining:
...     print("It's raining!")        # nothing happens
...
>>>
```

## Using else

## elif

If we have multiple conditions to check, we might write our code like this:

```python
print("Hello!")
word = input("Enter something: ")

if word == "hi":
    print("Hi to you too!")
else:
    if word == "hello":
        print("Hello hello!")
    else:
        if word == "howdy":
            print("Howdyyyy!")
        else:
            if word == "hey":
                print("Hey hey hey!")
            else:
                if word == "gday m8":
                    print("Gday 4 u 2!")
                else:
                    print("I don't know what", word, "means.")
```

It might look a bit messy — but this problem can be neatly solved using`elif`.

```elif``` is actually short for “else if.”

```python
print("Hello!")
word = input("Enter something: ")  # input

if word == "hi":
    print("Hi to you too!")
elif word == "hello":
    print("Hello hello!")
elif word == "howdy":
    print("Howdyyyy!")
elif word == "hey":
    print("Hey hey hey!")
elif word == "gday m8":
    print("Gday 4 u 2!")
else:
    print("I don't know what", word, "means.")
```


````python
print("Hello!")
word = input("Enter something: ")

if word == "hi":
    print("Hi to you too!")
elif word == "hello":
    print("Hello hello!")
elif word == "howdy":
    print("Howdyyyy!")
elif word == "hey":
    print("Hey hey hey!")
elif word == "gday m8":
    print("Gday 4 u 2!")
else:
    print("I don't know what", word, "means.")
````

````python
# Q3
arr = [1,12,14,21,23,36,43,48,56,78,89,102,126,147,168,171,199,222,256,348]
n = len(arr)

if n % 2 == 1:
    median = arr[n // 2]
else:
    mid1 = arr[n // 2 - 1]
    mid2 = arr[n // 2]
    median = (mid1 + mid2) / 2

print(median)
````
